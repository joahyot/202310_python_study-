#### 데이터프레임 인덱스 설정 및 제거

```python
# set_index: 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
# reset_index: 기존의 행 인덱스를 제거하고 인덱스를 데이터 열로 추가
```

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3,5), 2)]).T,
                              columns = ["c1", "c2", "c3", "c4"])
df1
```

```python
df2 = df1.set_index("c1")
df2
```

```python
# 다른 열을 인덱스로 지정하면 기존의 인덱스는 사라진다. 
df2.set_index("c2")
```

```python
df2.reset_index()
```

```python
# reset_index 메서드를 호출할 때 인수 drop=True로 설정하면 인덱스 열을 보통의 자료열로 올리는 것이 아니고 그냥 버리게 된다.

df2.reset_index(drop=True)
```

```python
### 연습문제 4.5.1

# 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터 프레일을 다음과 같이 만든다.

# 1. 학생 이름을 나타내는 열을 포함시키지 않고 데이터 프레임 df_score1을 생성한 후, 
# df_sort.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다.
# reset_index 명령으로 인덱스 열을 명령으로 일반 데이터열로 바꾸어 데이터 프레임 df_score2를 만든다.

# 2. 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터 프레임 df_score2에 set_index 명령을 적용하여
# 다시 학생 이름을 나타내는 열을 인덱스로 변경한다. 
```

```python
np.random.seed(0)
df_score1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3,5), 2)]).T,
                              columns = ["학생", "국어", "영어", "수학"])
df_score1
```

```python
df_score2 = df_score1.set_index("학생")
df_score2
```

```python
df_score2.reset_index()
```

#### 다중인덱스

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5,4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C1", "C2", "C1", "C2"]])
df3
```

```python
# columns 객체의 names 속성에 리스트를 넣어서 지정한다. 
df3.columns.names = ["Cidx1", "Cidx2"]
df3
```

```python
# 마찬가지로 데이터프레인을 생성할 때 index인 수에 리스트의 리스트 형태로 인덱스를 넣으면 다중 (행) 인덱스를 가진다. 
# 행 인덱스들의 이름 지정은 index 객체의 names 속성에 리스트를 넣어서 지정한다. 
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

#### 행 인덱스와 열 인덱스 교환

```python
# stack이나 unstack 메서드를 쓰면 열 인덱스를 행 인덱스로 바꾸거나 반대로 행 인덱스를 열 인덱스로 바꿀 수 있다.

df4.stack("Cidx1")
```

```python
df4.stack(1)
```

```python
df4.unstack("Ridx2")
```

```python
df4.unstack(0)
```
